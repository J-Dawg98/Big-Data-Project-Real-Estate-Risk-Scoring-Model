```python
# Excluding OPM remarks & Assessor Remarks, because of high missing values
# Drop one or more columns by name
df = df.drop(columns=["OPM remarks", "Assessor Remarks", "Non Use Code", "Location"])
```

```python
# Cleaning Dataframe
df_clean = df[(df["Sale Amount"].notna()) & (df["Town"].notna())]

# Convertir explicitement en datetime
df["Date Recorded"] = pd.to_datetime(df["Date Recorded"], errors="coerce", format="%m/%d/%Y")

#print(df["Date Recorded"].dtype)

df_clean["Date Recorded"] = pd.to_datetime(df_clean["Date Recorded"], errors='coerce')
df_clean["Year"] = df_clean["Date Recorded"].dt.year
```

```python
# Filling Na's of cleaned dataframe with 0
df_clean["Year"] = df_clean["Year"].fillna(0)
df_clean["Property Type"] = df_clean["Property Type"].fillna('Unknown')
df_clean["Residential Type"] = df_clean["Residential Type"].fillna('Unknown')
df_clean["Address"] = df_clean["Address"].fillna('Unknown')
df_clean["Date Recorded"] = df_clean["Date Recorded"].fillna('Unknown')
```

```python
# Example: Turn comma numbers into integers
df_clean["Year"] = df_clean["Year"].fillna(0).astype(int)

# Add Column 'State'
df_clean['State'] = 'Connecticut'
```

#### Interpretation:
- Top 3 by year that we had the most transactions: 2004, 2020 and 2003. Than the rest is pretty constant after 2005
  - 2003: real estate boom -> between 2001 and 2006
  - 2004: real estate boom (Easy credit, low interest rates, issuing risky mortgages to unqualified buyers (called subprime loans))
  - 2020: interest rate low, remote work (people move to countryside, so the transactions of houses increase), recovery plan, etc.
- Top 3 by year that we had the less transactions: 2011, 2008 and 2010. All the worst is pretty constant
  - 2007-2008: we had the subprime crisis, so it makes sens.
  - 2009-2010: we had the re-activity
  - 2010-2011: we had the waiting period which people wasn't confident to sell or buy and a lot of transactions was postponed
  - 2011: It fails so in 2011, it decreases.
- So, **economics activity** (**credit**, **low interest rates**, **mortgages**, **crisis**) are the indicator we should look closely to have more transactions
